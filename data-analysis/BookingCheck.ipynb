```python
import pandas as pd

file_path = "/Users/lottebecking/Library/CloudStorage/OneDrive-PurmerendseScholengroep/Uva/sem_6/Bestanden van Noel Lilliestielke - DreamSpace x UvA/Data collection/DS data/bookings.xlsx"
bookings_df = pd.read_excel(file_path)

bookings_df.head()
```

```python
from IPython.display import display, HTML

today = pd.Timestamp.today().normalize()
end_date = today + pd.Timedelta(days=30)

mask = (
    bookings_df["location"].str.contains("schiphol", case=False, na=False)
    & bookings_df["date"].between(today, end_date)
)

schiphol_next_30_days = bookings_df.loc[mask].sort_values(["date", "start_time"])

if schiphol_next_30_days.empty:
    display(HTML(f"<p style='color:#999;font-style:italic;'>No Schiphol bookings found between {today.date()} and {end_date.date()}.</p>"))
else:
    cols = ["date", "start_time", "end_time", "school", "location", "status", "workshop"]
    df_display = schiphol_next_30_days[cols].copy()
    df_display["date"] = df_display["date"].dt.strftime("%a %d %b %Y")

    styled = (
        df_display.style
        .set_caption(f"Schiphol bookings — {today.strftime('%d %b %Y')} to {end_date.strftime('%d %b %Y')}")
        .set_table_styles([
            {"selector": "caption", "props": [
                ("font-size", "15px"), ("font-weight", "bold"),
                ("text-align", "left"), ("padding-bottom", "8px"), ("color", "#333")
            ]},
            {"selector": "thead th", "props": [
                ("background-color", "#4A90D9"), ("color", "white"),
                ("font-size", "13px"), ("padding", "8px 12px"), ("text-align", "left")
            ]},
            {"selector": "tbody td", "props": [
                ("padding", "7px 12px"), ("font-size", "13px")
            ]},
            {"selector": "tbody tr:hover td", "props": [
                ("background-color", "#888 !important")
            ]},
            {"selector": "table", "props": [
                ("border-collapse", "collapse"), ("width", "100%")
            ]},
        ])
        .applymap(
            lambda v: "color: green; font-weight: bold;" if str(v).lower() == "confirmed"
            else ("color: orange;" if str(v).lower() == "pending" else ""),
            subset=["status"]
        )
        .hide(axis="index")
    )
    display(styled)
```
